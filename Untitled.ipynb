```python
def plot_mfcc(file):
    for fn in file:
        fns,sr=librosa.load(str(fn))
        mfcc = librosa.feature.mfcc(fns, n_mfcc=40)
        delta2_mfcc = librosa.feature.delta(mfcc, order=2)
        lab=str(fn)
        label=lab.split('/')[1]
        plt.figure(figsize=(12, 4))
        dp.specshow(delta2_mfcc)
        plt.ylabel('MFCC coeffs')
        plt.xlabel('Time')
        plt.title(label+' MFCC')
        plt.colorbar()
        plt.tight_layout()
plot_mfcc(file_p)
```

# Audio  sample graph

```python
def plot_audio_signal(file):
    fig = plt.figure(figsize=(14, 8))
    for i, fn in enumerate(file):
        wav,sr= librosa.load(fn)
        lab=str(fn)
        label=lab.split('/')[1]
        ax = fig.add_subplot(3,1,i+1)
        ax.set_title('Raw wave plot of ' + label)
        ax.set_ylabel('Amplitude')
        ax.plot(wav)
    fig.tight_layout()

plot_audio_signal(file_p)
```

# Coverting audio into MFCC

```python
def wav2mfcc(file_path,max_length=1500):
    wave, sr = librosa.load(file_path, mono=True,sr=20000,)
    mfcc = librosa.feature.mfcc(wave, sr=sr,n_mfcc=40)
    mfcc = np.pad(mfcc, ((0,0), (0, max_length-len(mfcc[0]))), mode='constant', constant_values=0)
    return mfcc
```

```python
data_root=pathlib.Path('data')
all_audio_path = list(data_root.glob('*/*'))
```

# Turn the converted data into a numerical  value as an array

```python
# split the data from the label into different form
x,y=[],[]
def train_set_split():
    data_root=pathlib.Path('data')
    all_image_paths = list(data_root.glob('*/*'))
    for wav in all_image_paths:
        wave=wav2mfcc(wav)
        label = str(wav)
        x.append(wave)
        y.append(label.split('/')[-2])
train_set_split()
```

```python
x_train=np.array(x)
x_train[3]
```

```python
x_t,y_tt=[],[]
def test_set_split():
    data_root=pathlib.Path('test')
    all_image_paths = list(data_root.glob('*/*'))
    for wav in all_image_paths:
        wave=wav2mfcc(wav)
        label = str(wav)
        x_t.append(wave)
        y_tt.append(label.split('/')[-2])
test_set_split()
```

```python
x_test=np.array(x_t)
```

```python
print("the shape of train value :" ,x_train.shape)
print("the shape of train value :" ,x_test.shape)
```

```python
lb = LabelEncoder()
y_t=lb.fit_transform(y)
y_tt=lb.fit_transform(y_tt)
```

```python
y_train=keras.utils.to_categorical(y_t)
y_test=keras.utils.to_categorical(y_tt)
```

```python
n_class=y_train.shape[1]
y_train.shape
y_test.shape
```

# Training the dataset after spliting

```python
callbacks = [ModelCheckpoint('voice_recognition_best_model_{epoch:02d}.hdf5', save_best_only=True)]
```

```python
model = keras.Sequential([
   
    keras.layers.LSTM(200,input_shape=(40,1500),return_sequences=True),
    
    keras.layers.LSTM(100,return_sequences=True),
    keras.layers.LSTM(50,return_sequences=True),
    keras.layers.BatchNormalization(),
    
    keras.layers.LSTM(35,return_sequences=True),
    keras.layers.LSTM(20,return_sequences=True),
    keras.layers.BatchNormalization(),
    
    keras.layers.LSTM(18,return_sequences=True),
    keras.layers.Conv1D(16,activation=tf.nn.relu,kernel_size=3,padding='same'),
    keras.layers.MaxPool1D(pool_size=2),
    keras.layers.BatchNormalization(),
    
    keras.layers.Flatten(),
    keras.layers.Dense(60, activation=tf.nn.relu),
    keras.layers.Dense(5, activation=tf.nn.relu),
    keras.layers.Dense(n_class, activation=tf.nn.softmax)
])
model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.002), 
              loss=tf.losses.categorical_crossentropy,
              metrics=['accuracy'])
history=model.fit(x_train, y_train, epochs=200,batch_size=32,validation_data=(x_test,y_test),callbacks=callbacks)
```
